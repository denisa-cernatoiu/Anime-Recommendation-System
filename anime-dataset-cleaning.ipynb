```python
import pandas as pd
df = pd.read_csv('/kaggle/input/anilist-anime-dataset/anilist_anime_data_complete.csv')
df.info()
```

```python
df.head(5)
```

```python
new_df = df[['title_romaji', 'title_english', 'meanScore', 'description', 'genres', 'format', 'status', 'episodes', 'studios', 'coverImage_large']].copy()
new_df.head()
```

```python
new_df.shape
```

```python
new_df.info()
```

```python
print(f'The min score is : {new_df['meanScore'].min()} and the max score is {new_df['meanScore'].max()}')
```

```python
# setting the score between 0 and 10
# removing animes with a score below 5 to prevent the model from generating poor recommendations

new_df['meanScore'] = new_df['meanScore'] / 10
new_df = new_df[new_df['meanScore'] > 5]
new_df = new_df.rename(columns={'meanScore': 'rating'})
```

```python
new_df = new_df[~new_df['genres'].str.contains('Hentai')]
new_df['genres'] = new_df['genres'].str.replace(r'[\[\]"]', '', regex=True)
```

```python
new_df['episodes'] = new_df['episodes'].fillna(0)
new_df['episodes'] = new_df['episodes'].astype(int)
```

```python
import re
def get_studios(text):
    res = re.findall(r'"name":\s*"([^"]+)"',text)
    return ", ".join(res)


new_df['studios'] = new_df['studios'].apply(lambda x: get_studios(x))
```

```python
new_df.sort_values(by='rating',ascending=False, inplace=True, ignore_index=True)
```

```python
new_df.head(10)
```

```python
new_df['title_english'].mask(new_df['title_english'].isna() == True, '', inplace=True)
new_df['description'].mask(new_df['description'].isna() == True, 'No description yet', inplace=True)
new_df['format'].mask(new_df['format'].isna() == True, 'Unknown', inplace=True)
```

```python
new_df.info()
```

```python
new_df.isna().sum()
```

```python
new_df["description"] = new_df["description"].apply(lambda x: re.sub(r"<.*?>", "", x))
```
