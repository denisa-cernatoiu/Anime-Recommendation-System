<h2 style="color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px;">Data Preparation Strategy</h2>
<p> Steps </p>
<ul style="line-height: 1.8;">
<li>trimming the dataset: keeping only the specific columns that actually influence our results </li>
<li>standardizing ratings</li>
<li>cleaning up formats</li>
<li>handling missing info</li>
<li>filtering the results</li>
<li>saving the dataset</li>


```python
new_df = df[['title_romaji', 'title_english', 'meanScore', 'description', 'genres', 'format', 'episodes', 'studios', 'coverImage_large']].copy()
new_df.head()
```

```python
new_df.shape
```

```python
new_df.info()
```

```python
print(f'The min score is : {new_df['meanScore'].min()} and the max score is {new_df['meanScore'].max()}')
```

```python
# setting the score between 0 and 10
# removing animes with a score below 5 to prevent the model from generating poor recommendations

new_df['meanScore'] = new_df['meanScore'] / 10
new_df = new_df[new_df['meanScore'] > 5]
new_df = new_df.rename(columns={'meanScore': 'rating'})
```

```python
new_df = new_df[~new_df['genres'].str.contains('Hentai')]
new_df['genres'] = new_df['genres'].str.replace(r'[\[\]"]', '', regex=True)
```

```python
import re
def get_studios(text):
    res = re.findall(r'"name":\s*"([^"]+)"',text)
    return ", ".join(res)


new_df['studios'] = new_df['studios'].apply(lambda x: get_studios(x))
```

```python
new_df.sort_values(by='rating',ascending=False, inplace=True, ignore_index=True)
```

```python
new_df['title_english'] = new_df['title_english'].fillna(new_df['title_romaji'])
new_df['description'] = new_df['description'].fillna('No description yet')
new_df['format'] = new_df['format'].fillna('Unknown')
```

```python
new_df["description"] = new_df["description"].apply(lambda x: re.sub(r"<.*?>", "", x))
```
